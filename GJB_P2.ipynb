#### Implementa el siguiente circuito cuántico con Qiskit y simúlalo. Recuerda que, por defecto, el estado inicial es $\ket{000}$, ¿cuál es el estado final? (mide los 3 qubits). El objetivo es averiguar qué hace el circuito cuántico.

#### Importacion de Librerias

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Operator
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.primitives import StatevectorSampler
import numpy as np
```

```python
circ = QuantumCircuit(3, 3) #Se crea un circuito con 3 cubits
circ.h(2)       # Se aplica H al tercer cubit
circ.cx(1, 2)   # Se aplica CNOT 
circ.tdg(2)     # Se aplica la compuerta T Dagger
circ.cx(0, 2)   # Se aplica CNOT 
circ.t(2)       # Se aplica la Compuerta T
circ.cx(1,2)    # Se vuelve a aplicar CNOT con el cubit 1 y 2
circ.tdg(2)     # Se aplica la compuerta T Dagger al cubit 2
circ.cx(0,2)    # Se aplica CNOT
circ.t(1)       # Se aplica T al cubit 1  
circ.t(2)       # Se aplica T al cubit 2
circ.cx(0,1)    # Se aplica la compuerta CNOT
circ.h(2)       # Se aplica la compuerta H al cubit 2
circ.t(0)       # Se aplica la compuerta T
circ.tdg(1)     # Se aplica la compuerta T Dagger            
circ.cx(0,1)    # Se aplica la compuerta CNOT         
circ.barrier()  # Se aplica una division   

circ.measure([0,1,2], [0,1,2]) # Se miden los cubits

display(circ.draw('mpl'))   # mostrar el circuito
```

<a id="1.1"></a>
### 1.1.2 Simulacion del circuito

```python
from qiskit.providers.basic_provider import BasicSimulator
simulator_basic = BasicSimulator()
job = simulator_basic.run(circ, shots=1024) ##Colapsa a un estado
result = job.result()
result
counts = result.get_counts()
counts
plot_histogram(counts, legend=["Estado 0"])
```

# 2. Ejerciocio 2

 Implementa la compuerta $SWAP$ pero utilizando su equivalencia de tres compuertas de 2 qubits (tercer diagrama en [esta referencia](https://pennylane.ai/qml/glossary/what-is-a-swap-gate/)) en código de `spinqit`.

<a id="1.1"></a>
## 2.1. Circuito 

```python
from spinqit import get_basic_simulator, get_compiler, BasicSimulatorConfig
from spinqit import Circuit, CX, draw
from spinqit import X
from spinqit.utils import print_circuit

circ = Circuit()
q = circ.allocateQubits(2)            
circ << (CX, (q[0], q[1]))
circ << (CX, (q[1], q[0]))
circ << (CX, (q[0], q[1]))

print(print_circuit(circ))
print()
comp = get_compiler()
engine = get_basic_simulator()
exe = comp.compile(circ, 0)
draw(exe)

config = BasicSimulatorConfig()
config.configure_shots(1024)
result = engine.execute(exe, config)
print(result.counts)
print(result.probabilities)
print(result.states)
print()
```

#### ¿Que crees que hace el circuito ?

El circuito cuántico utiliza compuertas X, CCX (Toffoli) y CX (CNOT) para realizar operaciones lógicas. Las compuertas X inicializan los cúbits 0, 1 y 2 en el estado |1⟩, actuando como NOT. Las compuertas CCX funcionan como un AND cuántico, activando el cúbit objetivo solo si ambos cúbits de control son 1, mientras que las compuertas CX actúan como XOR, invirtiendo el cúbit objetivo si el cúbit de control es 1. Aplicando estos operadores a todas las combinaciones posibles de 4 bits (2^4 = 16 estados), el circuito podría utilizarse para detectar patrones específicos de bits de entrada, implementar una función lógica compleja, o filtrar ciertas condiciones en algoritmos cuánticos, por ejemplo, en el oráculo de un algoritmo de Grover, donde se marca un estado objetivo específico entre muchas posibles entradas.
